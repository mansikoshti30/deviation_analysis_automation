```python
import os
```

```python
def explore_folder(folder_path):
    """
    Explore the given folder and list all subfolders and files.
    """
    if not os.path.exists(folder_path):
        print(f"Path does not exist: {folder_path}")
        return

    print(f"\nğŸ“‚ Exploring: {folder_path}\n")

    for root, dirs, files in os.walk(folder_path):
        # Print current folder
        level = root.replace(folder_path, "").count(os.sep)
        indent = " " * 4 * level
        print(f"{indent}ğŸ“ {os.path.basename(root)}/")

        # Print subfolders
        sub_indent = " " * 4 * (level + 1)
        for d in dirs:
            print(f"{sub_indent}ğŸ“‚ {d}/")

        # Print files
        for f in files:
            print(f"{sub_indent}ğŸ“„ {f}")
```

```python
folder = r"D:/2_Analytics/6_plan_vs_actual/8_sept_outputs_3"
explore_folder(folder)
```
