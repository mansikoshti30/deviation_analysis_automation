```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import matplotlib.patches as mpatches

# --- Step 1: Load shapefiles ---
planned_and_done = gpd.read_file("D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/planned_and_done_excavation.shp")
unplanned_and_done = gpd.read_file("D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/unplanned_and_done_excavation.shp")
planned_and_not_done = gpd.read_file("D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/planned_and_not_done_excavation.shp")

# --- Step 2: Calculate total areas in hectares ---
planned_and_done_area = planned_and_done.geometry.area.sum() / 10000  # ha
unplanned_and_done_area = unplanned_and_done.geometry.area.sum() / 10000
planned_and_not_done_area = planned_and_not_done.geometry.area.sum() / 10000

print("Planned & Done (ha):", planned_and_done_area)
print("Unplanned & Done (ha):", unplanned_and_done_area)
print("Planned & Not Done (ha):", planned_and_not_done_area)

# --- Step 3: Pie Chart ---
plt.figure(figsize=(6,6))
areas = [planned_and_done_area, unplanned_and_done_area, planned_and_not_done_area]
labels = ["Planned & Done", "Unplanned & Done", "Planned & Not Done"]
colors = ["#4CAF50", "#FF5722", "#2196F3"]

plt.pie(areas, labels=labels, autopct="%.1f%%", startangle=90, colors=colors)
plt.title("Excavation Area Distribution (Ha)")
plt.show()

# --- Step 4: Venn Diagram (Conceptual, not exact areas) ---
plt.figure(figsize=(6,6))
venn = venn2(subsets=(planned_and_not_done_area, unplanned_and_done_area, planned_and_done_area),
             set_labels=("Planned", "Done"))
plt.title("Planned vs Done Excavation")
plt.show()

# --- Step 5: Bar Chart ---
plt.figure(figsize=(6,6))
plt.bar(["Planned & Done", "Unplanned & Done", "Planned & Not Done"], areas, color=colors)
plt.ylabel("Area (Ha)")
plt.title("Excavation Area by Category")
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_excavation_chart(
    planned_done_file, unplanned_done_file, planned_notdone_file,
    labels=None, colors=None, figsize=(4,4)
):
    """
    Reads shapefiles, computes areas (in hectares), 
    plots a donut chart and returns a dataframe of results.
    """

    # === Read shapefiles ===
    gdf_planned_done = gpd.read_file(planned_done_file)
    gdf_unplanned_done = gpd.read_file(unplanned_done_file)
    gdf_planned_notdone = gpd.read_file(planned_notdone_file)

    # === Compute areas (in hectares) ===
    area_planned_done = gdf_planned_done.geometry.area.sum() / 10000
    area_unplanned_done = gdf_unplanned_done.geometry.area.sum() / 10000
    area_planned_notdone = gdf_planned_notdone.geometry.area.sum() / 10000

    # === Prepare data ===
    if labels is None:
        labels = ["Planned & active", "Unplanned & active", "Planned & inactive"]
    if colors is None:
        colors = ["#0DA21C", "#E71D0E", "#E6D00E"]  

    sizes = [area_planned_done, area_unplanned_done, area_planned_notdone]

    # === Plot donut chart ===
    fig, ax = plt.subplots(figsize=figsize)
    wedges, texts, autotexts = ax.pie(
        sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        wedgeprops=dict(width=0.4)
    )

    # === Reposition autopct text inside donut ===
    for i, a in enumerate(autotexts):
        ang = (wedges[i].theta2 + wedges[i].theta1) / 2.0
        x = np.cos(np.deg2rad(ang))
        y = np.sin(np.deg2rad(ang))
        a.set_position((0.8*x, 0.8*y))   
        a.set_color("white")
        a.set_weight("bold")

    ax.set(aspect="equal")
    plt.show()

    # === Create DataFrame with values ===
    df = pd.DataFrame({
        "Category": labels,
        "Area_ha": sizes
    })

    return df


df_result = plot_excavation_chart(
    "D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/planned_and_done_excavation.shp",
    "D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/unplanned_and_done_excavation.shp",
    "D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/planned_and_not_done_excavation.shp"
)

print(df_result)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

def append_summary_and_chart(
    csv_file, planned_done_file, unplanned_done_file, planned_notdone_file,
    output_csv=None, chart_file=None
):
    """
    Appends summary table (areas + percentages) to the right side of an existing CSV 
    and saves a pie chart as PNG.
    """

    # === Read shapefiles ===
    gdf_planned_done = gpd.read_file(planned_done_file)
    gdf_unplanned_done = gpd.read_file(unplanned_done_file)
    gdf_planned_notdone = gpd.read_file(planned_notdone_file)

    # === Compute areas in hectares ===
    area_planned_done = gdf_planned_done.geometry.area.sum() / 10000
    area_unplanned_done = gdf_unplanned_done.geometry.area.sum() / 10000
    area_planned_notdone = gdf_planned_notdone.geometry.area.sum() / 10000

    # === Summary DataFrame ===
    labels = ["Planned & active", "Unplanned & active", "Planned & inactive"]
    sizes = [area_planned_done, area_unplanned_done, area_planned_notdone]
    total = sum(sizes)
    percentages = [(s/total)*100 if total > 0 else 0 for s in sizes]

    summary_df = pd.DataFrame({
        "Category": labels,
        "Area_ha": sizes,
        "Percentage": percentages
    })

    # === Read existing CSV ===
    df = pd.read_csv(csv_file)

    # === Append summary table to right side ===
    # reset index for horizontal concat
    df_out = pd.concat([df, summary_df.reindex(df.index)], axis=1)

    # === Save output CSV ===
    if output_csv is None:
        output_csv = csv_file.replace(".csv", "_with_summary.csv")
    df_out.to_csv(output_csv, index=False)

    # === Create donut chart ===
    fig, ax = plt.subplots(figsize=(6,6))
    wedges, texts, autotexts = ax.pie(
        sizes, labels=labels, autopct='%1.1f%%', startangle=90,
        colors=["#0DA21C", "#E71D0E", "#E6D00E"],
        wedgeprops=dict(width=0.4)
    )

    # Move % labels inside
    for i, a in enumerate(autotexts):
        ang = (wedges[i].theta2 + wedges[i].theta1) / 2.0
        x, y = np.cos(np.deg2rad(ang)), np.sin(np.deg2rad(ang))
        a.set_position((0.8*x, 0.8*y))
        a.set_color("white")
        a.set_weight("bold")

    ax.set(aspect="equal")

    # === Save pie chart ===
    if chart_file is None:
        chart_file = csv_file.replace(".csv", "_summary_chart.png")
    plt.savefig(chart_file, dpi=300, bbox_inches="tight")
    plt.close()

    return output_csv, chart_file


csv_out, chart_out = append_summary_and_chart(
    csv_file= "D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/merged_summary.csv",
    planned_done_file= "D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/planned_and_done_excavation.shp",
    unplanned_done_file="D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/unplanned_and_done_excavation.shp",
    planned_notdone_file="D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/planned_and_not_done_excavation.shp"
)

print("Updated CSV saved at:", csv_out)
print("Pie chart saved at:", chart_out)
```

```python
import geopandas as gpd
import pandas as pd

def append_summary_and_chart_excel(
    csv_file, planned_done_file, unplanned_done_file, planned_notdone_file,
    output_excel=None
):
    # === Read shapefiles ===
    gdf_planned_done = gpd.read_file(planned_done_file)
    gdf_unplanned_done = gpd.read_file(unplanned_done_file)
    gdf_planned_notdone = gpd.read_file(planned_notdone_file)

    # === Compute areas in hectares ===
    area_planned_done = gdf_planned_done.geometry.area.sum() / 10000
    area_unplanned_done = gdf_unplanned_done.geometry.area.sum() / 10000
    area_planned_notdone = gdf_planned_notdone.geometry.area.sum() / 10000

    # === Summary DataFrame ===
    labels = ["Planned & Done", "Unplanned Done", "Plan'd NotDone"]
    sizes = [area_planned_done, area_unplanned_done, area_planned_notdone]
    total = sum(sizes)
    percentages = [(s/total)*100 if total > 0 else 0 for s in sizes]

    summary_df = pd.DataFrame({
        "Category": labels,
        "Area_ha": sizes,
        "Percentage": percentages
    })

    # === Read CSV ===
    df = pd.read_csv(csv_file)

    # === Output Excel ===
    if output_excel is None:
        output_excel = csv_file.replace(".csv", "_with_chart.xlsx")

    with pd.ExcelWriter(output_excel, engine="xlsxwriter") as writer:
        # Write data
        df.to_excel(writer, sheet_name="Data", index=False)
        summary_df.to_excel(writer, sheet_name="Summary", index=False)

        workbook  = writer.book
        worksheet = writer.sheets["Summary"]

        # === Create Pie Chart ===
        chart = workbook.add_chart({"type": "pie"})

        chart.add_series({
            "name":       "Area Distribution",
            "categories": ["Summary", 1, 0, len(summary_df), 0],  # Category column
            "values":     ["Summary", 1, 1, len(summary_df), 1],  # Area_ha column
            "data_labels": {"percentage": True},
        })

        chart.set_title({"name": "Planned vs Unplanned Areas"})
        chart.set_style(10)

        worksheet.insert_chart("E2", chart)

    return output_excel


excel_out = append_summary_and_chart_excel(
    csv_file="D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/merged_summary.csv",
    planned_done_file="D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/planned_and_done_excavation.shp",
    unplanned_done_file="D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/unplanned_and_done_excavation.shp",
    planned_notdone_file="D:/2_Analytics/6_plan_vs_actual/demo_test_2d_output_2/shape_file_outputs/planned_and_not_done_excavation.shp"
)

print("Excel with chart saved at:", excel_out)
```
